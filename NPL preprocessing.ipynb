```python
# importing the necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
import requests
import re
```

```python
# importing the excel with the urls
url_doc = pd.read_excel('C:/Users/acer/Desktop/Datasets/request_links_npl.xlsx')
url_doc.head()
```

```python
urls = url_doc['URLS '].to_list()
```

```python
# initiating a dictionary to store the json
json_responses = {}
```

```python
# looping through the url and fetching the json
for idx, url in enumerate(urls):
    try:
        response = requests.get(url)
        if response.status_code != 200:
            print(f"Failed to fetch the data from URL{idx+1}: {response.status_code}")
            continue
        json_responses[f"json_{idx+1}"] = response.json()
    except Exception as e:
        print(f"An error occured for URL{idx+1}: {e}")

# printing the data
for key, item in json_responses.items():
    print(f"{key}: {item}")
        
```
